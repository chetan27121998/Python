### IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### READING & CLEANING OF DATA AND FORMATTING IT PROPERLY

```python
data = pd.read_csv("carsdata.csv")
```

```python
data
```

```python
data.rename(columns={'Make':"Brand"},inplace=True) #renaming done of (make) to Brand
```

```python
data.head()
```

### FIND ALL THE NULL VALUES IN DATASET. IF THERE IS ANY NULL VALUE IN ANY COLUMN THEN FILL IT WITH THE MEAN OF THAT COLUMN

```python
data.isnull().sum()
```

```python
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace = True)
data
```

### INFORMATION OF THE FILE

### CHECK WHAT ARE THE DIFFERENT TYPE OF BRAND ARE THERE PRESENT IN OUR DATASET.AND WHAT IS THE COUNT (OCCURRENCE) OF EACH BRAND IN DATA?

```python
data.value_counts("Brand")
```

### FINDING THE TOTAL COUNT OF BRAND | WHICH BRAND HAS MORE DEMAND IN RESALE MARKET

```python
plt.figure(figsize=(19,19))
data['Brand'].value_counts().plot(kind='bar')
plt.title("MODEL")
plt.ylabel("NO of model")
plt.xlabel("BRAND")
plt.xticks(rotation=90)
plt.show()
```

### TOTAL NUMBER OF CAR IN LIST WITH CAR MODEL

```python
plt.figure(figsize=(10,10))
data['Model'].value_counts().nlargest(10).plot(kind='barh') # pandas plot as there is a lot of data so i took the .nlargest function
plt.title("Car Brand")
plt.ylabel("Brand")
plt.xlabel("count")
plt.xticks(rotation=90)
plt.show()
```
